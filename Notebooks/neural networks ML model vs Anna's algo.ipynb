```python
# import libraries
import pandas as pd
import numpy as np
from pathlib import Path
import talib
import hvplot.pandas
```

```python
#Import data

btc_data = pd.read_csv(Path("BTC-AUD.csv"),
    index_col='Date', 
    infer_datetime_format=True, 
    parse_dates=True)

# Print the historical data
btc_data
```

```python
# Delete Close and using adj close
del btc_data['Close']
btc_data
```

```python

btc_data ["MA15"] = talib.MA(btc_data["Adj Close"], timeperiod = 15)
btc_data ["MA30"] = talib.MA(btc_data["Adj Close"], timeperiod = 30)
btc_data ["MA60"] = talib.MA(btc_data["Adj Close"], timeperiod = 60)
btc_data ["VOL_60"] = talib.MA(btc_data["Volume"], timeperiod = 60)
btc_data ["RSI30"] = talib.RSI(btc_data["Adj Close"], timeperiod = 30)
btc_data 
```

```python
# Initialize the new Signal column to hold the trading signal
algo1_btc_data = btc_data.copy()
algo1_btc_data['Profit/Loss'] = 0
algo1_btc_data['Enter/Exit'] = 0.0
algo1_btc_data = algo1_btc_data.dropna()
algo1_btc_data
```

```python
# Generate the trading signal -1 or 1,
# where 1 is the close price is higher than the maximum of MA15, MA30, MA60, and Volume is higher than the VOL_60 by 20%. The take-profit strategy is the RSI30 ABOVE 85. The stop-loss strategy is 10% loss.
in_position = False
initial_fund = 50000
btc_amount = 0
enter_price = 0

for index, row in algo1_btc_data.iterrows():
    if in_position == True and (row["Adj Close"] < enter_price * 0.9 or row["RSI30"] > 85):
        in_position = False
        algo1_btc_data.at[index, "Enter/Exit"] = -1
        algo1_btc_data.at[index, "Profit/Loss"] = (row["Adj Close"] - enter_price) * btc_amount
        initial_fund = initial_fund + (row["Adj Close"] - enter_price) * btc_amount
    elif in_position == False and row["Adj Close"] > max(row["MA15"], row["MA30"], row["MA60"]) and row["Volume"] > row["VOL_60"]*1.2 and row["Adj Close"]>row["Open"]:
        in_position = True
        algo1_btc_data.at[index, "Enter/Exit"] = 1
        enter_price = row["Adj Close"]
        btc_amount = initial_fund / enter_price
    pass
    
        

algo1_btc_data
```

```python
algo1_btc_data["Accumulated Earnings"] = algo1_btc_data["Profit/Loss"].cumsum()+ 50000
```

```python
initial_fund = 50000
entry_price = algo1_btc_data[algo1_btc_data["Enter/Exit"] == 1]
first_entry_price = entry_price["Adj Close"].iloc[0]
entry_index = entry_price.index[0]
btc = initial_fund/first_entry_price
entry_index
algo1_btc_data["HOLDING"] = 0
algo1_btc_data.loc[algo1_btc_data.index >= entry_index, "HOLDING"] = btc * algo1_btc_data["Adj Close"]
```

```python
algo1_btc_data
```

```python
# Visualize exit position relative to close price
exit = algo1_btc_data[algo1_btc_data['Enter/Exit'] == -1.0]['Adj Close'].hvplot.scatter(
    color='yellow',
    marker='v',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize entry position relative to close price
entry = algo1_btc_data[algo1_btc_data['Enter/Exit'] == 1.0]['Adj Close'].hvplot.scatter(
    color='purple',
    marker='^',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize close price for the investment
security_close = algo1_btc_data[['Adj Close']].hvplot(
    line_color='lightgray',
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize moving averages
moving_avgs = algo1_btc_data[['MA15', 'MA30','MA60']].hvplot(
    ylabel='Price in $',
    width=1000,
    height=400
)

# Create the overlay plot
entry_exit_plot = security_close * moving_avgs * entry * exit

# Show the plot with a title
entry_exit_plot.opts(
    title="BTC - MA5, MA15, MA30 Entry and Exit Points"
)
```

```python
# Visualize exit position relative to total portfolio value
exit = algo1_btc_data[algo1_btc_data['Enter/Exit'] == -1.0]['Accumulated Earnings'].hvplot.scatter(
    color='red',
    marker='v',
    legend=False,
    ylabel='Accumulated Earnings',
    width=1000,
    height=400
)

# Visualize entry position relative to total portfolio value
entry = algo1_btc_data[algo1_btc_data['Enter/Exit'] == 1.0]['Accumulated Earnings'].hvplot.scatter(
    color='green',
    legend=False,
    marker='^',
    ylabel='Accumulated Earning',
    width=1000,
    height=400
)

# Visualize the value of the total portfolio
total_portfolio_value = algo1_btc_data[['Accumulated Earnings']].hvplot(
    line_color='Blue',
    ylabel='Accumulated Earning',
    xlabel='Date',
    width=1000,
    height=400
)

#compare with the holding strategy.
holding_close = algo1_btc_data[['HOLDING']].hvplot(
    line_color='Gray',
    ylabel='Accumulated Earning',
    width=1000,
    height=400
)
    
# Visualize the close price
btc_close = algo1_btc_data[['Adj Close']].hvplot(
    line_color='orange',
    ylabel='Price in $',
    width=1000,
    height=400
)

# Overlay the plots
portfolio_entry_exit_plot = btc_close * total_portfolio_value * entry * exit
portfolio_entry_exit_plot.opts(
    title="BTC Algorithm - Total Portfolio Value",
    yformatter='%.0f'
)
```

```python
# Overlay the plots, holding strategy
portfolio_entry_exit_plot = btc_close * total_portfolio_value * entry * exit*holding_close
portfolio_entry_exit_plot.opts(
    title="BTC Algorithm - Total Portfolio Value",
    yformatter='%.0f'
)
```

```python
# Machine Learning -Neural Networks
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

ml_strategy = btc_data.copy()
```

```python
# Calculate the daily returns using the closing prices and the pct_change function
ml_strategy["actual_returns"] = ml_strategy["Adj Close"].pct_change()

# Drop all NaN values from the DataFrame
ml_strategy = ml_strategy.dropna()

# Review the DataFrame
display(ml_strategy.head())
display(ml_strategy.tail())
```

```python
# Create a new column in the ml_strategy called signal setting its value to zero.
ml_strategy["signal"] = 0.0
# Create the signal to buy
ml_strategy.loc[(ml_strategy["actual_returns"] >= 0), "signal"] = 1
# Create the signal to sell
ml_strategy.loc[(ml_strategy["actual_returns"] < 0), "signal"] = 0
# Copy the new signal column to a new Series called y.
y = ml_strategy["signal"].copy()
y
```

```python
# Assign a copy of the ml_strategy to a new DataFrame called X
X = ml_strategy.drop(columns = ["signal","actual_returns"])

# Display sample data
display(X.head())
display(X.tail())
```

```python
# Imports 
from pandas.tseries.offsets import DateOffset

# Select the start of the training period
training_begin = X.index.min()

# Display the training begin date
print(training_begin)
```

```python
# Select the ending period for the training data with an offset of 3 months
training_end = X.index.min() + DateOffset(months=60)

# Display the training end date
print(training_end)
```

```python
# Generate the X_train and y_train DataFrames
X_train = X.loc[training_begin:training_end]
y_train = y.loc[training_begin:training_end]
# Display sample data
X_train.head()
```

```python
# Generate the X_test and y_test DataFrames
X_test = X.loc[training_end:]
y_test = y.loc[training_end:]

# Display sample data
X_test.head()
```

```python
# Create training and testing datasets using train_teat_split
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create the StandardScaler instance
X_scaler = StandardScaler()
```

```python
# Fit the scaler to the features training dataset
X_scaler.fit(X_train)
```

```python
# Scale both the training and testing data from the features dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the the number of inputs to the model
number_inputs = 10

# Define the number of hidden nodes for the first hidden layer
hidden_nodes_layer1 = 30

# Define the number of hidden nodes for the second hidden layer
hidden_nodes_layer2 = 20

# Create the Sequential model instance
nn = Sequential()

# Add the first Dense layer specifying the number of inputs, the number of hidden nodes, and the activation function
nn.add(Dense(units=hidden_nodes_layer1, input_dim=number_inputs, activation="relu"))

# Add the second Dense layer specifying the number of hidden nodes and the activation function
nn.add(Dense(units=hidden_nodes_layer2, activation="relu"))

# Add the output layer to the model specifying the number of output neurons and activation function
nn.add(Dense(1, activation='sigmoid'))
```

```python
# Display the Sequential model summary
nn.summary()
```

```python
# Compile the Sequential model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model using 100 epochs and the training data
model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled, y_test, verbose=2)

# Display the evaluation results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Set the model's file path
file_path = Path("neuro.h5")

# Export your model to an HDF5 file
nn.save(file_path)
```
